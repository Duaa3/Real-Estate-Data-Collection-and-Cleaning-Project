# Enhanced Data Visualizations 
This notebook provides more in-depth plots to explore the cleaned rental listings dataset.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Load Final Cleaned Data

```python
df = pd.read_csv('final_cleaned_listings.csv')

display(df.head())
```

## 3. Histogram of Area

```python
plt.figure(figsize=(8,4))
f = df['area'].hist(bins=40)
plt.title('Distribution of Listing Area (sqm)')
plt.xlabel('Area (sqm)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## 4. Histogram of Price

```python
plt.figure(figsize=(8,4))
df['price'].hist(bins=40)
plt.title('Distribution of Listing Price (OMR)')
plt.xlabel('Price (OMR)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## 5. Top 10 Locations by Average Price per sqm

```python
avg_price_loc = df.groupby('location')['price_per_sqm'].mean().nlargest(10)
plt.figure(figsize=(8,4))
avg_price_loc.plot(kind='barh')
plt.title('Top 10 Locations by Average Price per sqm')
plt.xlabel('Avg Price per sqm')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

## 6. Scatter: Price vs. Bedrooms

```python
plt.figure(figsize=(8,6))
plt.scatter(df['bedrooms'], df['price'], alpha=0.5)
plt.title('Listing Price vs. Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Price (OMR)')
plt.tight_layout()
plt.show()
```

## 7. Boxplot: Area by Bedrooms

```python
plt.figure(figsize=(8,6))
data_to_plot = [df[df['bedrooms']==b]['area'] for b in sorted(df['bedrooms'].unique())]
plt.boxplot(data_to_plot, labels=sorted(df['bedrooms'].unique()), vert=False)
plt.title('Boxplot of Area by Number of Bedrooms')
plt.xlabel('Area (sqm)')
plt.ylabel('Bedrooms')
plt.tight_layout()
plt.show()
```
