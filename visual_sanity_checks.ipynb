# Imports & Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_combined_listings_second_cleaned.csv')
```

# Histogram of Price

```python
plt.figure()
plt.hist(df['price'], bins=30)
plt.title('Histogram of Price')
plt.xlabel('Price (OMR)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Boxplot of Price

```python
plt.figure()
plt.boxplot(df['price'], vert=False)
plt.title('Boxplot of Price')
plt.xlabel('Price (OMR)')
plt.tight_layout()
plt.show()
```

# Histogram of Area

```python
plt.figure()
plt.hist(df['area'], bins=30)
plt.title('Histogram of Area')
plt.xlabel('Area (sqft)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Boxplot of Area

```python
plt.figure()
plt.boxplot(df['area'], vert=False)
plt.title('Boxplot of Area')
plt.xlabel('Area (sqft)')
plt.tight_layout()
plt.show()
```

# Histogram of log_price

```python
plt.figure()
plt.hist(df['log_price'], bins=30)
plt.title('Histogram of log_price')
plt.xlabel('Log Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Histogram of log_area

```python
plt.figure()
plt.hist(df['log_area'], bins=30)
plt.title('Histogram of log_area')
plt.xlabel('Log Area')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Bar Chart of Top 20 Locations

```python
top_locs = df['location'].value_counts().head(20)
plt.figure(figsize=(8,6))
plt.barh(top_locs.index[::-1], top_locs.values[::-1])
plt.title('Top 20 Locations by Listing Count')
plt.xlabel('Count')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```
